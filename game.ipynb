# EDA Analysis

```python
avg = df.groupby("Region")["Gaming_Revenue_BillionUSD"].mean().reset_index()

px.bar(avg,
       x="Region",
       y="Gaming_Revenue_BillionUSD",
       color="Region",
       title="Average Gaming Revenue by Region").show()
```

```python
fig = px.scatter(df, x="Year" ,y="Esports_Revenue_MillionUSD",color="Gaming_Revenue_BillionUSD",title="global_gaming_esports_2010_2025")
fig.show()
```

```python
year = df.groupby("Year")["Gaming_Revenue_BillionUSD"].mean().reset_index()
px.line(year,
        x="Year",
        y="Gaming_Revenue_BillionUSD",
        markers=True,
        title="Average Gaming Revenue by Year").show()
```

```python
sns.boxplot(df,y="Gaming_Revenue_BillionUSD")
plt.show()
```

```python
fig = px.violin(df,y="Esports_Revenue_MillionUSD",title="violin graph")
fig.show()
```

```python
fig = px.scatter_3d(df,
                    x="Mobile_Gaming_Share",
                    y="Esports_PrizePool_MillionUSD",
                    z="Gaming_Companies_Count",
                    color="Gaming_Companies_Count",
                    hover_name="Country",
                    title="3D Space Exploration Analysis")
fig.show()
```

```python
platform = df["Top_Platform"].value_counts()
plt.figure(figsize=(6,6))
plt.pie(platform, labels=platform.index, autopct="%1.1f%%", startangle=90)
plt.title("Top Platform Distribution")
plt.show()
```

```python
genre = df["Top_Genre"].value_counts().reset_index()
genre.columns = ["Genre", "Count"]
fig = px.pie(
genre,
names="Genre",
values="Count",
hole=0.5,  # Makes it a donut chart
title="Top Game Genre Distribution")
fig.update_traces(textposition='inside', textinfo='percent+label')
fig.show()
```

```python
# 5. Histogram
sns.histplot(df,x="Pro_Players_Count",bins=20, kde=True)
plt.show()
```
