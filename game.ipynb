# EDA Analysis

### Bar Graph

```python
avg = df.groupby("Region")["Gaming_Revenue_BillionUSD"].mean().reset_index()

px.bar(avg,
       x="Region",
       y="Gaming_Revenue_BillionUSD",
       color="Region",
       title="<b>Average Gaming Revenue by Region</b>").show()
```

### Scatter graph

```python
df["Esports_Revenue_MillionUSD"] = df["Esports_Revenue_MillionUSD"] *1000

fig = px.scatter(
    df,
    x="Year",
    y="Esports_Revenue_MillionUSD",
    title="<b>Esports_Revenue (MillionUSD) According to Yearly.</b>",
    color="Esports_Revenue_MillionUSD"
)

fig.show()
```

### Box plot Graph

```python
num_cols = df.select_dtypes(include="number").columns
scaled_df = df[num_cols].copy()

scaled_df = (scaled_df - scaled_df.min()) / (scaled_df.max() - scaled_df.min())

fig = px.box(
    scaled_df,
    y=scaled_df.columns,
    points="outliers",
    title="Box Plot of All Numerical Features"
)

fig.update_layout(
    template="plotly_white",
    height=700,
    width=1200,
    title_x=0.5,
    showlegend=False
)

fig.show()
```

### 3D Graph

```python
fig = px.scatter_3d(
    df,
    x="Mobile_Gaming_Share",
    y="Esports_PrizePool_MillionUSD",
    z="Gaming_Companies_Count",
    color="Gaming_Companies_Count",
    hover_name="Country",
    title=(
        "<b>3D Analysis of the Global Gaming Industry</b><br>"
        "<b>Mobile Gaming, Esports Prize Pools, and Gaming Companies</b><br>"
        "<sup>Exploring the relationship between mobile gaming market share, esports prize pools,and the number of gaming companies across countries.</br> "
        "Each point represents a country,while color intensity indicates the number of gaming companies.</sup>"
        )
)
fig.update_layout(
    title={
        "x": 0.5,
        "xanchor": "center",
        "font": {"size": 20}
    },
    width=1100,
    height=800,
    margin=dict(l=0, r=0, t=140, b=0), 
    scene=dict(
        xaxis_title="Mobile Gaming Share (%)",
        yaxis_title="Esports Prize Pool (Million USD)",
        zaxis_title="Gaming Companies Count"
    )
)
fig.show()
```

### Pie Graph

```python
import plotly.express as px

platform = df["Top_Platform"].value_counts().reset_index()
platform.columns = ["Top_Platform", "Count"]

fig = px.pie(
    platform,
    names="Top_Platform",
    values="Count",
    title=("<b>Distribution of Users According to Top Platforms</b>")
)

fig.update_traces(
    textinfo="percent+label",
    textposition="inside",
)
fig.update_layout(
    title={
        "xanchor": "left",
        "font": {"size": 20}}
)
             
    

fig.show()
```

### Donut Graph

```python
genre = df["Top_Genre"].value_counts().reset_index()
genre.columns = ["Genre", "Count"]

fig = px.pie(
   genre,
   names="Genre",
   values="Count",
   hole=0.5,  # Makes it a donut chart
   title=("<b>Top Game Genre Distribution</b>")
)

fig.update_traces(
        textposition='inside',
        textinfo='percent+label'),
fig.update_layout(
    title={
        "xanchor": "left",
        "font": {"size": 20}}
)
fig.show()
```

### Line Graph

```python
year = df.groupby("Year")["Gaming_Revenue_BillionUSD"].mean().reset_index()
fig = px.line(
    year,
    x="Year",
    y="Gaming_Revenue_BillionUSD",
    markers=True,
    title="<b>Average Gaming Revenue by Year</b>"
)

fig.show()
```
